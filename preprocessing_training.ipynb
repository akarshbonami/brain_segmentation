```python
import os
import numpy as np
import nibabel as nib
import tensorflow as tf
import matplotlib.pyplot as plt
from tensorflow.keras.layers import Conv2D, MaxPooling2D, Conv2DTranspose, concatenate, Input
from tensorflow.keras.models import Model
from sklearn.model_selection import train_test_split
```

##Load and Preprocess Data

```python
DATASET_PATH = "/kaggle/working/NFBS_Dataset"
IMG_SIZE = (128, 128)


def load_mri_data(dataset_path, img_size):
    images, masks = [], []

    folders = sorted(os.listdir(dataset_path))  # Limit to 10 folders
    for folder in folders:
        folder_path = os.path.join(dataset_path, folder)

        t1w_path = os.path.join(folder_path, f"sub-{folder}_ses-NFB3_T1w.nii.gz")
        mask_path = os.path.join(folder_path, f"sub-{folder}_ses-NFB3_T1w_brainmask.nii.gz")

        if os.path.exists(t1w_path) and os.path.exists(mask_path):
            # Load images
            t1w_img = nib.load(t1w_path).get_fdata()
            mask_img = nib.load(mask_path).get_fdata()

            # Resize images
            t1w_img = tf.image.resize_with_pad(t1w_img, img_size[0], img_size[1]).numpy()
            mask_img = tf.image.resize_with_pad(mask_img, img_size[0], img_size[1]).numpy()

            # Add new axis for channel (grayscale)
            images.append(t1w_img[..., np.newaxis])
            masks.append(mask_img[..., np.newaxis])

    return np.array(images), np.array(masks)

# Load dataset
images, masks = load_mri_data(DATASET_PATH, IMG_SIZE)
print(f"Loaded {len(images)} images and {len(masks)} masks")
```

```python
# split dataset
X_train, X_test, y_train, y_test = train_test_split(images, masks, test_size=0.2, random_state=42)
print(f"Training images: {X_train.shape}, {y_train.shape},Testing images: {X_test.shape}, {y_test.shape}")
```

```python
import tensorflow as tf
from tensorflow.keras import layers, models

def unet_3d(input_shape=(128, 128, 192, 1)):
  input = layers.Input(input_shape)

  # Encoder
  conv1 = layers.Conv3D(32, (3,3,3), activation="relu", padding="same")(input)
  conv1 = layers.Conv3D(32, (3,3,3), activation="relu", padding="same")(conv1)
  pool1 = layers.MaxPooling3D(pool_size=(2,2,2))(conv1)

  conv2 = layers.Conv3D(64, (3,3,3), activation="relu", padding="same")(pool1)
  conv2 = layers.Conv3D(64, (3,3,3), activation="relu", padding="same")(conv2)
  pool2 = layers.MaxPooling3D(pool_size=(2,2,2))(conv2)

  # Bottleneck
  conv3 = layers.Conv3D(128, (3,3,3), activation="relu", padding="same")(pool2)
  conv3 = layers.Conv3D(128, (3,3,3), activation="relu", padding="same")(conv3)

  # Decoder
  up4 = layers.Conv3DTranspose(64, (2,2,2), strides=(2,2,2), padding="same")(conv3)
  concat4 = layers.concatenate([up4, conv2])
  conv4 = layers.Conv3D(64, (3,3,3), activation="relu", padding="same")(concat4)
  conv4 = layers.Conv3D(64, (3,3,3), activation="relu", padding="same")(conv4)

  up5 = layers.Conv3DTranspose(32, (2,2,2), strides=(2,2,2), padding="same")(conv4)
  concat5 = layers.concatenate([up5, conv1])
  conv5 = layers.Conv3D(32, (3,3,3), activation="relu", padding="same")(concat5)
  conv5 = layers.Conv3D(32, (3,3,3), activation="relu", padding="same")(conv5)

  outputs = layers.Conv3D(1, (1,1,1), activation="sigmoid")(conv5)

  model = models.Model(inputs=input, outputs=outputs)
  return model
```

##Create the model

```python
model = unet_3d(input_shape=(128, 128, 192, 1))
model.compile(optimizer="adam", loss="binary_crossentropy", metrics=["accuracy"])
model.summary()
```

```python
history = model.fit(
    X_train, y_train,
    validation_data = (X_test, y_test),
    epochs = 10,
    batch_size = 1
)
```

```python
tf.keras.backend.clear_session()
```

```python
loss, acc = model.evaluate(X_test, y_test, batch_size=4)
print(f"Test Accuracy: {acc:.4f}")
```

## 3D U-net + Resnset Architecture

```python
import tensorflow as tf
from tensorflow.keras import layers, models

def residual_block_3d(x, filters, kernel_size=(3,3,3), strides=(1,1,1)):
    """3D Residual Block with channel matching"""
    shortcut = x  # Preserve input for skip connection

    x = layers.Conv3D(filters, kernel_size, padding="same", strides=strides)(x)
    x = layers.BatchNormalization()(x)
    x = layers.ReLU()(x)

    x = layers.Conv3D(filters, kernel_size, padding="same", strides=strides)(x)
    x = layers.BatchNormalization()(x)

    # 🔹 **Fix: Apply 1x1x1 Conv if Channels Mismatch**
    if shortcut.shape[-1] != filters:
        shortcut = layers.Conv3D(filters, (1,1,1), padding="same")(shortcut)

    # Skip connection (adding input to output)
    x = layers.Add()([x, shortcut])
    x = layers.ReLU()(x)

    return x
```

```python
import tensorflow as tf
from tensorflow.keras import layers, models

def residual_block_3d(x, filters, kernel_size=(3,3,3), strides=(1,1,1)):
    """3D Residual Block with channel matching"""
    shortcut = x  # Preserve input for skip connection

    x = layers.Conv3D(filters, kernel_size, padding="same", strides=strides)(x)
    x = layers.BatchNormalization()(x)
    x = layers.ReLU()(x)

    x = layers.Conv3D(filters, kernel_size, padding="same", strides=strides)(x)
    x = layers.BatchNormalization()(x)

    # 🔹 **Fix: Apply 1x1x1 Conv if Channels Mismatch**
    if shortcut.shape[-1] != filters:
        shortcut = layers.Conv3D(filters, (1,1,1), padding="same")(shortcut)

    # Skip connection (adding input to output)
    x = layers.Add()([x, shortcut])
    x = layers.ReLU()(x)

    return x
```

```python
def resnet_3d_unet(input_shape=(128, 128, 192, 1)):
    inputs = layers.Input(input_shape)

    # === Encoder (with residual blocks) ===
    conv1 = residual_block_3d(inputs, 32)
    pool1 = layers.MaxPooling3D(pool_size=(2,2,2))(conv1)

    conv2 = residual_block_3d(pool1, 64)
    pool2 = layers.MaxPooling3D(pool_size=(2,2,2))(conv2)

    conv3 = residual_block_3d(pool2, 128)
    pool3 = layers.MaxPooling3D(pool_size=(2,2,2))(conv3)

    # === Bottleneck ===
    conv4 = residual_block_3d(pool3, 256)

    # === Decoder ===
    up5 = layers.Conv3DTranspose(128, (2,2,2), strides=(2,2,2), padding="same")(conv4)
    concat5 = layers.concatenate([up5, conv3])
    conv5 = residual_block_3d(concat5, 128)

    up6 = layers.Conv3DTranspose(64, (2,2,2), strides=(2,2,2), padding="same")(conv5)
    concat6 = layers.concatenate([up6, conv2])
    conv6 = residual_block_3d(concat6, 64)

    up7 = layers.Conv3DTranspose(32, (2,2,2), strides=(2,2,2), padding="same")(conv6)
    concat7 = layers.concatenate([up7, conv1])
    conv7 = residual_block_3d(concat7, 32)

    # === Output ===
    outputs = layers.Conv3D(1, (1,1,1), activation="sigmoid")(conv7)

    model = models.Model(inputs=inputs, outputs=outputs)
    return model
```

```python
model2 = resnet_3d_unet()
model2.compile(optimizer="adam", loss="binary_crossentropy", metrics=["accuracy"])
model2.summary()
```

```python
history = model2.fit(
    X_train, y_train,
    validation_data=(X_test, y_test),
    epochs=7,
    batch_size=1
)
```

```python
loss, acc = model2.evaluate(X_test, y_test, batch_size=4)
print(f"Test Accuracy: {acc:.4f}")
```
